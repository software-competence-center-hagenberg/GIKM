# Comparing Matlab Results with Python Results

```python
import numpy as np
import pandas as pd
from scipy.io import loadmat
import matplotlib.pyplot as plt
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
def plot_metric_comparison(metric, df1, df2, title, filename):
    # Extract values for the specific metric from both DataFrames
    value_df1 = df1.loc[df1['Metric'] == metric, 'Value'].values[0]
    value_df2 = df2.loc[df2['Metric'] == metric, 'Value'].values[0]

    # Create a bar plot comparing the two values
    plt.figure(figsize=(6, 4))
    plt.bar(['python', 'matlab'], [value_df1, value_df2], color=['blue', 'green'])
    
    # Adding labels and title
    plt.title(title)
    plt.ylabel('Value')
    plt.ylim(0, max(value_df1, value_df2) * 1.1)  # Set y-limit slightly higher than the max value

    # Save the plot to a file
    plt.tight_layout()
    #plt.savefig(f"{filename}.png")
    #plt.close()

```

## MNIST

```python
mnist_csv = "./accuracy_metrics_mnist.csv"
```

```python
acc = 0.98699999999999999
max_modeling_error = 0.0044211705765663245
```

```python
mnist_csv_df = pd.read_csv(mnist_csv)
mnist_mat_df = pd.DataFrame({
    'Metric': ['Max Modelling Error', 'Accuracy'],
    'Value': [max_modeling_error, acc]
})
```

```python
mnist_csv_df
```

```python
mnist_mat_df
```

```python

# Plot "Max_Modeling_Error" comparison
plot_metric_comparison("Max Modelling Error", mnist_csv_df, mnist_mat_df, "Max Modeling Error: python vs matlab", "mnist_max_modeling_error_comparison")

# Plot "Accuracy" comparison
plot_metric_comparison("Accuracy", mnist_csv_df, mnist_mat_df, "Accuracy: python vs matlab", "mnist_accuracy_comparison")
```
